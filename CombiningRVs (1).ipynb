#### The graph shows the distributions of random variables $X$ and $Y$, where $X$ is normally distributed with mean 20 and standard deviation 7, and $Y$ is normally distributed with mean 2 and standard deviation 4.

We can use the following notation to describe this:

$X \sim N(\mu = 20, \sigma^2=49)$

$Y \sim N(\mu = 2, \sigma^2=16)$

```python
import numpy as np
import matplotlib.pyplot as plt
X= np.random.normal(loc=20, scale=7, size=10000000)
Y = np.random.normal(loc=2, scale=4, size=10000000)
plt.hist(X, bins = 100, label='X', alpha=.5, color='red')
plt.hist(Y, bins = 100, label='Y', alpha=.5, color='blue')
plt.legend(loc='upper right')
plt.show
```

The next graph shows the distribution of $X$ and $X+30$. Notice that the mean shifts up 30, but the spread (standard deviation) stays the same.

```python
# New cell
Z=X+30
plt.hist(X, bins = 100, label='X', alpha=.5, color='red')
#plt.hist(Y, bins = 100, label='Y', alpha=.5)
plt.hist(Z, bins = 100, label='Z=X+30', alpha=.4, color='yellow')
plt.legend(loc='upper right')
plt.show
```

The next graph shows the distribution of $X$ and $\frac{1}{2}X+30$. Notice that the mean shifts up 30, and the spread decreases (by a factor of $\frac{1}{2}$)

```python
# New cell
Z=(1/2)*X+30
plt.hist(X, bins = 100, label='X', alpha=.5, color='red')
#plt.hist(Y, bins = 100, label='Y', alpha=.5)
plt.hist(Z, bins = 100, label='Z=(1/2)X+30', alpha=.4, color='yellow')
plt.legend(loc='upper right')
plt.show
```

The next graph shows the distribution of $X$, $Y$ and $X+Y$. Notice that the mean shifts by the sum of the means of $X$ and $Y$ (20 and 2), and the spread of the resulting random variable is larger than the spread of each individual random variable.

```python
# New cell
Z=X+Y
plt.hist(X, bins = 100, label='X', alpha=.5, color='red')
plt.hist(Y, bins = 100, label='Y', alpha=.5, color='blue')
plt.hist(Z, bins = 100, label='Z=X+Y', alpha=.4, color='yellow')
plt.legend(loc='upper right')
plt.show
```

The next graph shows the distribution of $X$, $Y$ and $X-Y$. Notice that the mean shifts by the difference of the means of $X$ and $Y$ (20 and 2), and the spread of the resulting random variable is still larger than the spread of each individual random variable.

```python
# New cell
Z=X-Y
plt.hist(X, bins = 100, label='X', alpha=.5, color='red')
plt.hist(Y, bins = 100, label='Y', alpha=.5, color='blue')
plt.hist(Z, bins = 100, label='Z=X-Y', alpha=.4, color='yellow')
plt.legend(loc='upper right')
plt.show
```
